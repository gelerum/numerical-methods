```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = lambda x: -(x-50)**2 + 2500
xs = np.arange(0, 101, 1)
ys = g(xs)

plt.xlabel('x')
plt.ylabel('g(x)')
plt.axhline(y=0, lw=1, color='k')
plt.axvline(x=0, lw=1, color='k')
plt.plot(xs, ys)
```

```python
N=101
```

```python
u = np.zeros((N, N))
for i in range(N):
    u[i][0] = ys[i]

print(u[0][0], u[100][0], u[50][0])
```
